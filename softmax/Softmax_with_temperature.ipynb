# Brief demonstation of softmax with temperature scaling

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Define softmax function

```python
def softmax_with_temp(x, T=1):
    """
    Lets define our temperature-scaled softmax function, with a default temperature of 1, which recovers the non-scaled distribution

    Args:
        x (np array): input vector to be scaled
        T (double): scaling temperature

    Returns:
        np array: softmax transformation
    """

    return np.exp(x/T)/sum(np.exp(x/T))
```

## Create random vector

```python
rng = np.random.default_rng()
```

```python
x = rng.integers(-100, 100, 20)
```

```python
x
```

## Apply softmax without and with temperature scaling

```python
softmax_with_temp(x)
```

```python
softmax_with_temp(x, 2)
```

## Visualise

```python
i = [str(i) for i in  range(0, len(x))]
```
